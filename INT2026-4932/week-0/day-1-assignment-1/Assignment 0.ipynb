```python
#inheritance
class A:
    def show(self):
        print("India Is best country")
class B(A):
    def display(self): 
        print("B class")
b=B()
b.show();
b.display()
```

```python
#polymorphism
class Dog:
    def sound(self):
        print("Bark")
class Cat:
    def sound(self):
        print("Meow")

for a in (Dog(),Cat()):
    a.sound()
```

```python
# multithreading
import threading
def show():
    print("Thread is running")
t = threading.Thread(target=show)
t.start()
t.join()
print("Main thread done")
```

```python
import threading
def task():
    for i in range(5):
        print("Task:", i)
t1 = threading.Thread(target=task)
t2 = threading.Thread(target=task)
t1.start()
t2.start()
t1.join()
t2.join()
```

```python
class Student:
    def __init__(self,n,m):
        self.name=n; self.marks=m
    def result(self):
        return "Pass" if self.marks>=40 else "Fail"
s=Student("A",55)
print(s.name, s.result())
```
